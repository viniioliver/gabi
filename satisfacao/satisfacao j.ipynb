```python
import pandas as pd #Para importar a biblioteca panda
import matplotlib.pyplot as plt #Para importar a biblioteca mat
import numpy as np #Numpy
df = pd.read_csv('satisfacao.csv', sep=';') #Ler arquivo CSV com nome alterado para somente satisfação
```

```python
q1 = np.percentile(df['Tempo de Espera'], 25) #Criado o calculo do primeiro quartil , segundo e terceiro. Exercicio pediu os 25% do primeiro quartil
q2 = np.percentile(df['Tempo de Espera'], 50) 
q3 = np.percentile(df['Tempo de Espera'], 75)
df_cincomaisrapidos = df.loc[df['Tempo de Espera'] <= q1, ['Cliente', 'Tempo de Espera']] #criado um df com os 25% dos clientes com menores tempos
#print(df_cincomaisrapidos)
df_cincomaisrapidos.plot(kind='bar', x='Cliente', y='Tempo de Espera', figsize=(10,5), legend=False) #criação do grafico (nao esquecer o.plot) 
plt.title('Tempo de espera')
plt.xlabel('Cliente')
plt.ylabel('Tempo em minutos')
plt.xticks(rotation=45, ha='right')  # gira os nomes para não sobrepor
plt.tight_layout()
plt.show()
```
